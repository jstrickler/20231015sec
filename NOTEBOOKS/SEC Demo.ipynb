```python
sales = pd.read_csv('../DATA/sales_records.csv', parse_dates=['Ship Date', 'Order Date'])
sales.columns = "region country itype channel priority orderdate orderid shipdate sold price cost".split()
sales
```

```python
recent_angola_sales  = sales[(sales['country'] == 'Angola') & (sales['shipdate'] > '20150101') ]
recent_angola_sales.to_csv('angola.csv')
recent_angola_sales.to_excel('angola.xlsx')
recent_angola_sales
```

```python
sales.total_sales = sales['sold'] * sales['price']
sales['total_cost'] = sales['sold'] * sales['cost']
sales['profit'] = sales['total_sales'] - sales['total_cost']
sales
```

```python
sales.profit.max()
```

```python
best_sales_idx = sales.profit.idxmax()
sales.loc[best_sales_idx]
```

```python
sales.drop(sales[sales.country == 'Panama'].index, inplace=True)
sales
```

```python
sales.groupby('region').mean('total_sales')
```
