# OLS Regression

## Standard imports

```python
# %load imports.py
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt  # matplotlib is package; pyplot is module
import matplotlib as mpl
%matplotlib inline
import seaborn as sns   
sns.set()
import statsmodels.api as sm  # modeling tools

# NOTE: demo copied from https://www.statology.org/ols-regression-python/ 

```

## Read the data

```python
data = pd.read_csv('../DATA/study_hours.csv')
data
```

```python
y = data['score']  # response
x = data['hours']  # predictor
```

```python
x = sm.add_constant(x)
x
```

```python
model = sm.OLS(y, x).fit()
```

```python
model.summary()
```
