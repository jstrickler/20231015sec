### Getting real memory use

```python
df.info(memory_usage='deep')
```

### Sum rows and column
Add a new column with sums of rows, and a new row with sums of columns

```python
df.loc['ALL ROWS'] = df.sum(numeric_only=True)
df['ALL COLUMNS'] = df.sum(numeric_only=True, axis=1)
df.tail()
```

### Only describe objects (text)

Note the high ratio of count to unique for *sex*, *embarked*, and *boat*. These are prime opportunities to save memory by using categoricals.

```python
titanic.describe(include='O')
```

### Count males vs females
We can use `.value_counts()` to count distinct values in the `sex` column

```python
passengers_by_sex = titanic.value_counts('sex', dropna=False)
passengers_by_sex
```

### Count passengers per embarcation point and make a bar chart

```python
passengers_by_emb = titanic.value_counts(['sex', 'embarked'])
plot = passengers_by_emb.plot(kind="bar")
plot.set_xticklabels('Southhampton-M Southhampton-F Cherbourg-M Cherbourg-F Queenstown-M Queenstown-F'.split())
```
